```python
#Reading the data
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
#converting the above two columns to the Boolean type
yes_no = {
    "Yes": True,
    "No": False
}
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no)
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)
```

```python
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].value_counts(dropna=False)
```

```python
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].value_counts(dropna=False)
```

```python
import numpy as np

# select the entire titles, which all appear on the first row, col 3-9, 
# and convert to a list:
movies = star_wars.iloc[0, 3:9].tolist()
print(movies)

# loop through the checkbox columns, mapping it to True or False 
# and renaming the column names: 

for i,col in zip(range(0,6), star_wars.columns[3:9]):
    yes_nan = {movies[i]:True, np.NaN: False}
    star_wars[col] = star_wars[col].map(yes_nan)
    new_col = "seen_" + str(i+1)
    star_wars = star_wars.rename(columns={col: new_col})
```

```python
star_wars.head(10)
```

```python
for col in star_wars.columns[3:9]:
    print(col)
```

```python
#cleaning the ranking columns
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
#renaming the ranking columns
for i, col in zip(range(0,6), star_wars.columns[9:15]):
    new_col = "ranking_"+ str(i+1)
    star_wars = star_wars.rename(columns={col: new_col})
    
```

```python
star_wars.columns[9:15]
```

```python
#Finding the highest Ranked movie
#computing the mean for ranking columns
%matplotlib inline
import matplotlib.pyplot as plt
mean_rank = star_wars[star_wars.columns[9:15]].mean()
mean_rank.plot.bar()
plt.title("Ranking")
plt.xlabel("Movies")
plt.ylabel("Average ranking")
plt.xticks(rotation=45)
plt.show()
```

We changed the ranking columns to numeric and changes their column name.
And we computed the average from each column to plot in the bargraph as shown above.
The fifth part of the movie is highly ranked on an average.

```python
#Finding the most viewed movie
sum_seen = star_wars[star_wars.columns[3:9]].sum()
sum_seen.plot.bar()
plt.title("Total Viewers of Movies")
plt.xlabel("total view")
plt.ylabel("movies")
plt.xticks(rotation=45)
plt.show()

    
```

As expected the movie part that had the highest rating has the highest number of viewers. The results are positively correlated with the ranking.

```python
#spliting the dataset into 2 based on the gender.
males = star_wars[star_wars["Gender"]== "Male"]
females = star_wars[star_wars["Gender"]== "Female"]
print(males.head(5))
print(females.head(5))
```

```python
#Finding the most viewed movie based on gender
#Finding the most viewed movie for males
sum_seen = males[males.columns[3:9]].sum()
sum_seen.plot.bar()
plt.title("Total Viewers of Movies for males")
plt.xlabel("total view")
plt.ylabel("movies")
plt.xticks(rotation=45)
plt.show()

#Finding the most viewed movie for females
sum_seen = females[females.columns[3:9]].sum()
sum_seen.plot.bar()
plt.title("Total Viewers of Movies for females")
plt.xlabel("total view")
plt.ylabel("movies")
plt.xticks(rotation=45)

plt.show()
```

From the above analysis we came to know that there are more male viewers in comparison to female.

```python
#Finding the highest ranked movie based on gender

#highest ranked movie for males
mean_rank = males[males.columns[9:15]].mean()
mean_rank.plot.bar()
plt.title("Ranking of Movies for Males")
plt.xlabel("Movies")
plt.ylabel("Average ranking")
plt.xticks(rotation=45)
plt.show()

#highest ranked movie for females
mean_rank = females[females.columns[9:15]].mean()
mean_rank.plot.bar()
plt.title("Ranking of Movies for Females")
plt.xlabel("Movies")
plt.ylabel("Average ranking")
plt.xticks(rotation=45)
plt.show()
```

There is no such difference on the ranking of movies. Both male and female
highly ranked the fifth part of the movie.
The study cocludes that the star wars part “The Empire Strikes Back” is the most highly ranked and mostly viewed movie.
